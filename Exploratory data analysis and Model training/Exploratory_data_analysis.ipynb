```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'B:\Ubiquitous Lab\Input data\Combined_data_with_var.csv')
#df = df.sample(frac=1)
df.drop(columns=['Unnamed: 0'], inplace=True)
df.info()

```

```python
df_lift = df[df['label']==0]
plt.xlabel("No.of datapoints")
plt.ylabel("Acceleration-Lift")
plt.plot(df_lift['x'])
plt.plot(df_lift['y'])
plt.plot(df_lift['z'])
plt.legend(['accel-x','accel-y','accel-z'])
```

```python
df_stairs = df[df['label']==1]
plt.xlabel("No.of datapoints")
plt.ylabel("Acceleration-Lift")
plt.plot(df_stairs['x'])
plt.plot(df_stairs['y'])
plt.plot(df_stairs['z'])
plt.legend(['accel-x','accel-y','accel-z'])
```

```python
var_x_staircase = df_lift['x'].var()
var_x_lift = df_stairs['x'].var()

var_y_staircase = df_lift['y'].var()
var_y_lift = df_stairs['y'].var()

var_z_staircase = df_lift['z'].var()
var_z_lift = df_stairs['z'].var()

Staircase_var = [var_x_staircase,var_y_staircase,var_z_staircase]
lift_var = [var_x_lift,var_y_lift,var_z_lift]

indices = np.arange(len(Staircase_var))

# Set the width of the bars
bar_width = 0.35

# Plotting the variances
plt.bar(indices, Staircase_var, width=bar_width, label='Staircase')
plt.bar(indices + bar_width, lift_var, width=bar_width, label='lift')

# Adding labels and title
plt.xlabel('acceleration')
plt.ylabel('Variance')
plt.title('Comparison of Variances')
plt.xticks(indices + bar_width / 2, ['x', 'y', 'z'])
plt.legend()

# Display the plot
plt.show()
```
